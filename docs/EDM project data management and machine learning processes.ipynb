## 4. Detecting student behavior and clustering processess using machine learning

```python
# Subset common important features across all sessions 
# by using an ensemble learning method fundamentally based on decision trees
dt1 = whole_data.copy()
data_common_important_features = mm.subset_important_features(dt1, 3, "common")
data_common_important_features[1] #Session 2
```

```python
# Subset important features from each session
# by using an ensemble learning method fundamentally based on decision trees
dt2 = whole_data.copy()
data_different_important_features = mm.subset_important_features(dt2, 5, "different")
data_different_important_features[2] #Session 3
```

```python
# Detect student behavior by fittng important log data and intermediate score into logistic regression
student_behavior_predictions = mm.machine_learning_model(data_different_important_features, 'LR')
student_behavior_predictions[2] # When Predicted_Y = 1, a student will get a notification from this system
```

```python
# Student behavior prediction accuracy
print(accuracy_score(student_behavior_predictions[1]['Y'], student_behavior_predictions[1]['Predicted_Y']))
print(accuracy_score(student_behavior_predictions[2]['Y'], student_behavior_predictions[2]['Predicted_Y']))
print(accuracy_score(student_behavior_predictions[3]['Y'], student_behavior_predictions[3]['Predicted_Y']))
print(accuracy_score(student_behavior_predictions[4]['Y'], student_behavior_predictions[4]['Predicted_Y']))
print(accuracy_score(student_behavior_predictions[5]['Y'], student_behavior_predictions[5]['Predicted_Y']))
```

```python
# Clustering students based on students' log activities and grades
dt3 = data_common_important_features.copy()
clustering_results = mm.kmean_clustering(dt3, 3, 3) # 3rd session and 3 clusters
clustering_results
```

```python
# Frequencies of each cluster
frequencies = {}
for item in clustering_results['group']:
    if item in frequencies:
        frequencies[item] += 1
    else:
        frequencies[item] = 1
print(frequencies)
```
