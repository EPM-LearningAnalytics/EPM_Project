```python
# load data
log_session = pd.read_csv('../EPM_dataset/Data/log_session_mere.csv')
log_session
```

## Altair

```python
# for each student:
# 1. their activity - log across sessions (bar plots)
base = alt.Chart(log_session, width=600, height=300).mark_bar().encode(
x = 'mouse_click_left',
y = 'session:N',
color = 'activity'
).transform_calculate(student = 'datum.student_id')

student_slider = alt.binding_range(min=1, max=115, step=1)
slider_selection = alt.selection_single(bind=student_slider, fields=['student'], name="Student ID")

filter_student = base.add_selection(
    slider_selection
).transform_filter(
    slider_selection
).properties(title="Student Filtering")

filter_student
```

## Bokeh

```python
# for each student:
# 1. their activity - log across sessions (bar plots)
from bokeh.plotting import figure, show
from bokeh.models import ColumnDataSource
from bokeh.palettes import Spectral8
```

```python
# choose only 1 student and 1 log activity
student1 = log_session[log_session['student_id'] == 7]
student1_mcl = student1[['session', 'activity','mouse_click_left']]
student1_mcl['session'] = student1_mcl['session'].astype(str)
student1_mcl.head()
```

```python
# transform the dataset to include activity as the columns, sessions as rows
stacked1 = student1_mcl.pivot(index = 'session', columns = 'activity', values = 'mouse_click_left')
#stacked1['session'] = stacked1['session'].apply(str)
activities = stacked1.columns.tolist()
stacked1
```

```python
source = ColumnDataSource(stacked1)
sessions = source.data['session']

p = figure(x_range = sessions, height=500, title='student 1 mouse_left_click')
```

```python
p.vbar_stack(activities, x = 'session', width = 0.9,
             color = Spectral8, source=source,
             legend_label=activities)
p.y_range.start = 0
p.x_range.range_padding = 0.1
p.xgrid.grid_line_color = None
p.axis.minor_tick_line_color = None
p.outline_line_color = None
p.legend.location = "top_right"
p.legend.orientation = "vertical"
```

```python
show(p)
```
